```python
country_name = 'India'  
country_data = df[df['Country Name'] == country_name]

# Bar chart
plt.figure(figsize=(50, 20))
plt.bar(country_data.columns[1:], country_data.values[0][1:], color='blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {country_name} over the years')
plt.show()
```

```python
plt.figure(figsize=(50,20))
plt.hist(country_data.values[0][1:], bins=20, color='skyblue')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Population distribution of {country_name}')
plt.show()
```
